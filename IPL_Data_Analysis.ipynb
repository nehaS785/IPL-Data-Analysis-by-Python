```python
# 1. What is the total number of runs scored by each batting team?
import matplotlib.pyplot as plt
import seaborn as sns

total_runs_by_team = df.groupby('batting_team')['total_runs'].sum()

# Plot the total runs by each batting team
plt.figure(figsize=(12, 6))
sns.barplot(x=total_runs_by_team.index, y=total_runs_by_team.values, palette='viridis')
plt.title("Total Runs by Each Batting Team")
plt.xticks(rotation=90)
plt.ylabel("Total Runs")
plt.show()
```

```python
# 2. How many wides have been bowled by each bowling team?
wides_by_team = df.groupby('bowling_team')['wide_runs'].sum()

plt.figure(figsize=(12, 6))
sns.barplot(x=wides_by_team.index, y=wides_by_team.values, palette='coolwarm')
plt.title("Total Wides Bowled by Each Bowling Team")
plt.xticks(rotation=90)
plt.ylabel("Total Wides")
plt.show()
```

```python
# 3. What is the average number of runs scored per inning by each team?
avg_runs_per_inning = df.groupby(['batting_team', 'inning'])['total_runs'].mean().groupby('batting_team').mean()
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_runs_per_inning.index, y=avg_runs_per_inning.values, palette='Set2')
plt.title("Average Runs Per Inning by Each Team")
plt.xticks(rotation=90)
plt.ylabel("Average Runs Per Inning")
plt.show()
```

```python
# 4. Which batsman has scored the highest number of runs?
top_batsman = df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_batsman.index, y=top_batsman.values)
plt.title("Top 10 Batsmen with Highest Runs")
plt.xticks(rotation=90)
plt.ylabel("Total Runs")
plt.show()
```

```python
# 5. How many dismissals have occurred of each dismissal kind?
dismissals_by_kind = df['dismissal_kind'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=dismissals_by_kind.index, y=dismissals_by_kind.values, palette='Paired')
plt.title("Number of Dismissals by Dismissal Kind")
plt.xticks(rotation=45)
plt.ylabel("Number of Dismissals")
plt.show()
```

```python
#  6. What is the total runs scored by each over?
runs_per_over = df.groupby('over')['total_runs'].sum()

# Plot the total runs scored by each over
plt.figure(figsize=(12, 6))
sns.lineplot(x=runs_per_over.index, y=runs_per_over.values, marker='o', color='blue')
plt.title("Total Runs Scored by Each Over")
plt.xlabel("Over")
plt.ylabel("Total Runs")
plt.grid(True)
plt.show()
```

```python
# 7. Which bowler has taken the most wickets?
top_bowlers = df[df['player_dismissed'].notna()].groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_bowlers.index, y=top_bowlers.values)
plt.title("Top 10 Bowlers with Most Wickets")
plt.xticks(rotation=90)
plt.ylabel("Wickets")
plt.show()
```

```python
# 8. How many super overs have been played in total?
total_super_overs = df['is_super_over'].sum()
plt.figure(figsize=(6, 4))
plt.pie([total_super_overs, df['match_id'].nunique() - total_super_overs], labels=['Super Overs', 'Regular Overs'], autopct='%1.1f%%')
plt.title("Proportion of Super Overs")
plt.show()
```

```python
# 9. What is the average number of extras (wide runs, bye runs, legbye runs, noball runs, penalty runs) conceded per inning by each bowling team?
extras_per_inning = df.groupby(['bowling_team', 'match_id', 'inning'])['extra_runs'].sum().groupby('bowling_team').mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=extras_per_inning.index.get_level_values(0), y=extras_per_inning.values)
plt.title("Average Extras Per Inning by Bowling Team")
plt.xticks(rotation=90)
plt.ylabel("Average Extras")
plt.show()
```

```python
# 10. How many runs have been scored by each batsman-bowler pair?
runs_by_batsman_bowler = df.groupby(['batsman', 'bowler'])['batsman_runs'].sum().reset_index()
top_batsman_bowler_pairs = runs_by_batsman_bowler.sort_values(by='batsman_runs', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_batsman_bowler_pairs.apply(lambda x: f"{x['batsman']} vs {x['bowler']}", axis=1), 
            y=top_batsman_bowler_pairs['batsman_runs'])
plt.title("Top 10 Batsman-Bowler Pairs with Most Runs")
plt.xticks(rotation=90)
plt.ylabel("Runs")
plt.show()
```

```python
# 11. What is the most common dismissal kind?
most_common_dismissal = df['dismissal_kind'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=most_common_dismissal.index, y=most_common_dismissal.values)
plt.title("Most Common Dismissal Kind")
plt.xticks(rotation=90)
plt.ylabel("Frequency")
plt.show()
```

```python
# 12. How many matches have been played by each batting team?
matches_by_team = df.groupby('batting_team')['match_id'].nunique()
plt.figure(figsize=(10, 6))
sns.barplot(x=matches_by_team.index, y=matches_by_team.values)
plt.title("Number of Matches Played by Each Batting Team")
plt.xticks(rotation=90)
plt.ylabel("Number of Matches")
plt.show()
```

```python
# 13. What is the average number of runs scored in each over?
runs_per_over = df.groupby(['over'])['total_runs'].sum()
avg_runs_per_over = runs_per_over / df.groupby(['over'])['match_id'].nunique()
plt.figure(figsize=(10, 6))
sns.lineplot(x=avg_runs_per_over.index, y=avg_runs_per_over.values, marker='o')
plt.title("Average Runs Scored in Each Over")
plt.xlabel("Over")
plt.ylabel("Average Runs")
plt.show()
```

```python
# 14. Which fielder has been involved in the most dismissals?
top_fielders = df['fielder'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_fielders.index, y=top_fielders.values)
plt.title("Top 10 Fielders Involved in Most Dismissals")
plt.xticks(rotation=90)
plt.ylabel("Dismissals")
plt.show()
```

```python
# 15. How does the number of runs scored vary with the over number?
runs_per_over = df.groupby('over')['total_runs'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=runs_per_over.index, y=runs_per_over.values, marker='o')
plt.title("Variation of Runs Scored with Over Number")
plt.xlabel("Over")
plt.ylabel("Total Runs")
plt.show()
```

```python
# 16. how is the relationship between the number of runs scored and the number of extras conceded?
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['total_runs'], y=df['extra_runs'])
plt.title("Relationship between Runs Scored and Extras Conceded")
plt.xlabel("Total Runs")
plt.ylabel("Extras")
plt.show()
```

```python
# 17. Are there any notable trends in the distribution of  runs scored by batting teams across innings?
runs_by_team_inning = df.groupby(['batting_team', 'inning'])['total_runs'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=runs_by_team_inning, x='inning', y='total_runs', hue='batting_team', marker='o')
plt.title("Distribution of Runs Scored by Batting Teams Across Innings")
plt.xlabel("Inning")
plt.ylabel("Total Runs")
plt.legend(title='Batting Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 18. Which batting team has the highest average runs per match?
total_runs_per_team_match = df.groupby(['batting_team', 'match_id'])['total_runs'].sum()
avg_runs_per_match = total_runs_per_team_match.groupby('batting_team').mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_runs_per_match.index, y=avg_runs_per_match.values)
plt.title("Average Runs Per Match by Batting Team")
plt.xticks(rotation=90)
plt.ylabel("Average Runs")
plt.show()
```

```python
# 19. What is the total runs scored by each batsman?
total_runs_by_batsman = df.groupby('batsman')['batsman_runs'].sum()
plt.figure(figsize=(12, 6))
top_batsmen = total_runs_by_batsman.sort_values(ascending=False).head(10)  # Optional: Top 10 batsmen for clarity
sns.barplot(x=top_batsmen.index, y=top_batsmen.values)
plt.title("Total Runs Scored by Each Batsman")
plt.xticks(rotation=90)
plt.ylabel("Total Runs")
plt.show()
```

```python
# 20. How does the average number of runs scored per inning vary across different bowling teams?
runs_against_bowling_team = df.groupby(['bowling_team', 'inning'])['total_runs'].sum().reset_index()
innings_count = df.groupby('bowling_team')['inning'].nunique()
avg_runs_by_bowling_team = runs_against_bowling_team.groupby('bowling_team')['total_runs'].mean()
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_runs_by_bowling_team.index, y=avg_runs_by_bowling_team.values)
plt.title("Average Runs Scored Per Inning Against Each Bowling Team")
plt.xticks(rotation=90)
plt.ylabel("Average Runs")
plt.show()
```

```python
# 21. how is the relationship between the number of wides and the number of no balls bowled?
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['wide_runs'], y=df['noball_runs'])
plt.title("Relationship between Wides and No Balls Bowled")
plt.xlabel("Wide Runs")
plt.ylabel("No Ball Runs")
plt.show()
```

```python
# 22. Which batting team has the highest number of centuries?
centuries = df[df['batsman_runs'] >= 100]
centuries_by_team = centuries.groupby('batting_team').size()
if not centuries_by_team.empty:
    plt.figure(figsize=(12, 6))
    sns.barplot(x=centuries_by_team.index, y=centuries_by_team.values, palette='viridis')
    plt.title("Number of Centuries by Each Batting Team")
    plt.xticks(rotation=90)
    plt.ylabel("Number of Centuries")
    plt.show()
else:
    print("No centuries found in the dataset.")
```

```python
# 23. How does the number of wides vary with the over number?
wides_per_over = df.groupby('over')['wide_runs'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=wides_per_over.index, y=wides_per_over.values, marker='o')
plt.title("Variation of Wides with Over Number")
plt.xlabel("Over")
plt.ylabel("Total Wides")
plt.show()
```

```python
# 24. What is the average number of runs scored per inning in matches with a super over?
super_over_matches = df[df['is_super_over'] == 1]
total_runs_super_over_inning = super_over_matches.groupby(['match_id', 'inning'])['total_runs'].sum()
avg_runs_super_over = total_runs_super_over_inning.groupby('inning').mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_runs_super_over.index, y=avg_runs_super_over.values)
plt.title("Average Runs Per Inning in Matches with Super Over")
plt.xlabel("Inning")
plt.ylabel("Average Runs")
plt.show()
```

```python
# 25. How does the number of extras vary with the over number?
total_extras_per_over = df.groupby('over')['extra_runs'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=total_extras_per_over.index, y=total_extras_per_over.values, marker='o')
plt.title("Variation of Extras with Over Number")
plt.xlabel("Over")
plt.ylabel("Total Extras")
plt.show()
```

```python
26. Which bowler has the highest economy rate?
total_runs_conceded = df.groupby('bowler')['total_runs'].sum()
total_overs_bowled = df.groupby('bowler')['over'].nunique()
economy_rate = total_runs_conceded / total_overs_bowled
highest_economy_rate_bowler = economy_rate.idxmax()
highest_economy_rate = economy_rate.max()

print(f"Bowler with the highest economy rate: {highest_economy_rate_bowler} ({highest_economy_rate:.2f})")
```

```python
# 27. How does the number of dismissals vary across different dismissal kinds?
dismissals_by_kind = df['dismissal_kind'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=dismissals_by_kind.index, y=dismissals_by_kind.values)
plt.title("Number of Dismissals by Dismissal Kind")
plt.xticks(rotation=90)
plt.ylabel("Number of Dismissals")
plt.show()
```

```python
# 28. What is the distribution of runs scored by each  batting team in super overs?
super_over_runs_by_team = df[df['is_super_over'] == 1].groupby('batting_team')['total_runs'].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=super_over_runs_by_team.index, y=super_over_runs_by_team.values)
plt.title("Distribution of Runs Scored by Each Batting Team in Super Overs")
plt.xticks(rotation=90)
plt.ylabel("Total Runs")
plt.show()
```
